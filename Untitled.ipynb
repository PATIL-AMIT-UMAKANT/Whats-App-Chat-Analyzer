```python
#importing pandas and Regular Expression
import re
import pandas as pd
```

```python
f=open('WhatsApp Chat with Happy engineers Day .txt','r',encoding='utf-8')
#utf:-
```

```python
data = f.read()
```

```python

print(type(data))
```

```python
print(data)
```

```python
pattern='\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{1,2}\s\w[aApPmM]\s-\s'
pattern
```

```python
messages = re.split(pattern, data)
messages 
```

```python
len(messages) 
```

```python
messages = re.split(pattern, data)[1:]
messages
```

```python
dates = re.findall(pattern, data)
dates
```

```python
df = pd.DataFrame({'user_message': messages, 'message_date': dates})
# convert message_date type
df
```

```python
df['message_date'] = pd.to_datetime(df['message_date'], format='%d/%m/%y, %H:%M %p - ')
df['message_date']
```

```python
df.rename(columns={'message_date': 'date'}, inplace=True)
```

```python
df.head()
```

```python
df.shape
```

```python
#19min
users = []
messages = []
for message in df['user_message']:
    entry = re.split('([\w\W]+?):\s', message)
    if entry[1:]: # user name 
        users.append(entry[1])
        messages.append(" ".join(entry[2:]))
    else:
        users.append('group_notification')
        messages.append(entry[0])

df['user'] = users
df['message'] = messages
df.drop(columns=['user_message'], inplace=True)


df.head()
```

```python
df['year'] = df['date'].dt.year
df
```

```python
df['month_num'] = df['date'].dt.month
df['month'] = df['date'].dt.month_name()
df
```

```python
df['only_date'] = df['date'].dt.date
df['hour'] = df['date'].dt.hour
df['minute'] = df['date'].dt.minute
```

```python
df['day'] = df['date'].dt.day
df['day_name'] = df['date'].dt.day_name()
df
```
